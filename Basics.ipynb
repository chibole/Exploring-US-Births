The first step is to carry out some data cleaning, making it user friendly and putting it in a list of lists for further exploration. 
 


```python
def read_csv(US_births_data):
    file_handle = open(US_births_data,'r')
    content = file_handle.read()
    split_string = content.split("\n")
    string_list= split_string[1:len(split_string)]
    final_list = []
    for item in string_list:
        int_fields = []
        string_fields = item.split(",")
        for value in string_fields:
            int_fields.append(int(value))
        final_list.append(int_fields)
    return(final_list)
cdc_list = read_csv("US_births_1994-2003_CDC_NCHS.csv")
cdc_list[:20]
```

For the sake of space the above output is trimmed to 20 elements but it runs all the way to December 31, 2003.

Notice that the dataset into a list of lists where each of the nested list contains integer values (not strings). The header row is also removed during the data cleanup.

## Calculating the Number of Births Each Month

```python
def month_births(monthly_births):
    births_per_month = {}
    for value in monthly_births:
        month = value[1]
        births = value[4]
        if month in births_per_month:
            births_per_month[month] += births
        else:
            births_per_month[month] = births
    return births_per_month
cdc_month_births = month_births(cdc_list)
cdc_month_births

```

The results above is the communicative births for all months from the year 1994 and 2003. 

## Calculating Number of Births Each Day of Week

```python
def dow_births(daily_births):
    births_per_day = {}
    for value in daily_births:
        day = value[3]
        births = value[4]
        if day in births_per_day:
            births_per_day[day] += births
        else:
            births_per_day[day] = births
    return births_per_day
cdc_daily_births = dow_births(cdc_list)
cdc_daily_births
    
    
```

## Creating a more General Function

Specifyses whether the column of interest is year, month or day of month or day of week

```python
def calc_counts(data, column):# column takes 0, 1, 2 or 3.
    births_dict = {}
    for value in data:
        col_value = value[column]
        births = value[4]
        if col_value in births_dict:
            births_dict[col_value] += births
        else:
            births_dict[col_value] = births
    return births_dict
    
cdc_year_births = calc_counts(cdc_list, 0)
cdc_month_births = calc_counts(cdc_list, 1)
cdc_dom_births = calc_counts(cdc_list, 2)
cdc_dow_births = calc_counts(cdc_list, 3) 
    
```

## Max and Min value Calculation

```python
def max_min(data, column):
    max_min_value = calc_counts(data, column)
    max_value = max(max_min_value.values())
    min_value = min(max_min_value.values())
       
    return max_value, min_value 
the_max = max_min(cdc_list, 2)
the_max
```

## Extracting the same value across years
Here is a function that extracts the same values across years and calculates the differences between consecutive values to show if number of births is increasing or decreasing.

For example, how did the number of births on Saturday change each year between 1994 and 2003?
In tackling this problem, I have to recall that Saturday is day 6, so  the function progressive() shows if consecutive births every Saturday (day == 6) are increasing or decreasing from  1994 to 2003.


```python
def progressive(births_data, column, num):
    clean_data = read_csv(births_data)
    saturday_births = []
    for value in clean_data:
        col_value = value[column]
        births = value[4]
        if col_value == num: # for day 6 of every month from 1994-2003
            saturday_births.append(int(births))
            if len(saturday_births) == 1: # the list has one element - nothing to compare
                print("No previous births to compare.")
            #compare last to second last birth in saturday_births list that increases by one element on eact iteration
            elif saturday_births[len(saturday_births)-1] > saturday_births[len(saturday_births)-2]:
                print("Births increased.")
            else:
                print("Births decreased.")
    return saturday_births
           
input_data = progressive("US_births_1994-2003_CDC_NCHS.csv", 3, 6)
```

## Function to merge data from two disparate sources.
In this problem, I have to combine the CDC data with the SSA data. The two have some years with overlapping data.

I first cleaned the data using the read_csv() function. Then I compared the first three elements in a tuple in list1 to the second list2. If none was in either list I addes it to the final merged list. If the tuple in list1  is the same as that of tuple in list2, I compared their last element (births). If the births are the same I picked list2 and added it to the final merged list (I could pick list1 as everything is the same), if births in list2 was higher than in list1, I added list2 to the final merged list, otherwise, I added list1 to the final merged list.
My reasoning is that if two sources have different births on the same day or month, the the highest birth is likely to be correct.

```python
def merge_sources(source1, source2):
    clean_data1 = read_csv(source1)
    clean_data2 = read_csv(source2)
    merged_list = []
    
    for item1 in clean_data1:
        data1 = tuple(item1[:3])
        if data1 not in clean_data2:
            merged_list.append(item1)
    for item2 in clean_data2:
        data2 = tuple(item2[:3])
        if data2 not in clean_data1:
            merged_list.append(item2)
    for data2 in clean_data1:
        if item2[4] == item1[4]:
            merged_list.append(item2)
        elif item2[4] > item1[4]:
            merged_list.append(item2)
        else:
            merged_list.append(item1)
            
    return merged_list  

sources = merge_sources("US_births_1994-2003_CDC_NCHS.csv", "US_births_2000-2014_SSA.csv")
sources[:20]
```
